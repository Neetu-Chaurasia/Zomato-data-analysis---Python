## Do high-rated restaurants always get more votes?

```python
sns.scatterplot(data=zomato,x='Total_Votes',y='Avg_Rating_Restaurant')
```

### Insights
No, **high-rated restaurants do not always get more votes**.

**Explanation:**

* Some restaurants have **high ratings but low votes**, which means fewer people reviewed them.
* Some restaurants have **many votes but average ratings**, showing they are popular but not always highly rated.
* Overall, there is **no strong direct relationship** between rating and number of votes.

**Conclusion:**

> A good rating does not guarantee more votes; popularity and visibility also matter.

### Inisghts

## Are delivery ratings generally higher than dining ratings?

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=zomato, x='Delivery_Rating', kde=True, label='Delivery', color='blue', alpha=0.5)
sns.histplot(data=zomato, x='Dining_Rating', kde=True, label='Dining', color='orange', alpha=0.5)
plt.title('Delivery vs Dining Ratings Distribution', fontsize=14)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### Insights
From the chart, **delivery ratings are generally slightly higher than dining ratings**. The delivery ratings (blue) are more concentrated around 4.0–4.2, while dining ratings (orange) have a wider spread and a noticeable peak at 4.0. This suggests customers tend to rate delivery experiences a bit higher on average than dining experiences.


# 5.Pricing and Value Analysis

## Analyze the relationship between an item's Average_Rating and its calculated Price_per_Vote.

```python
zomato.groupby('Item_Name').agg({
    'Average_Rating':'mean',
    'Price_per_Vote':'sum'
}).sort_values(by=['Average_Rating','Price_per_Vote'],ascending=False).head(10)
```

### Insights
* Items with **highest Average_Rating (4.65)** tend to have **higher Price_per_Vote** (773.73 for top combos, 257.91 for “Make your Own Combo”).
* Items with slightly lower ratings (4.45) have **moderate Price_per_Vote**, ranging from 176 to 412.
* **Conclusion:** Higher-rated items often have higher Price_per_Vote, but it’s **not strictly proportional**—some highly rated items like “Make your Own Combo” have lower Price_per_Vote.

In short, **better-rated items generally cost more per vote, but there are exceptions.**
